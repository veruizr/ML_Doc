# Árboles de Decisión y Bosques Aleatorios en Scikit-Learn
## 1. Introducción

### Árboles de Decisión (Decision trees)
Un **Árbol de Decisión** es un modelo de aprendizaje supervisado que toma decisiones basándose en reglas derivadas de los datos. Funciona como un diagrama de flujo:
- **Nodos**: Representan características o decisiones.
- **Ramas**: Representan los valores posibles de una característica.
- **Hojas**: Representan las clases o valores finales.

### Bosques Aleatorios (random forest)
Un **Bosque Aleatorio** es un conjunto organizado de Árboles de Decisión. Cada árbol se entrena con una muestra aleatoria de los datos y características, y las predicciones finales se obtienen combinando los resultados .


## 2. Árboles de Decisión

### Teoría Básica
Un árbol de decisión divide el espacio de características en regiones basadas en preguntas binarias (por ejemplo, "¿Es el valor de X > 5?"). Utiliza métricas como **Gini** o **Entropía** para determinar las mejores divisiones.


```python
### Ejemplo: Clasificación con el Dataset Iris

# Importar bibliotecas
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Cargar el dataset Iris
iris = load_iris()
X, y = iris.data, iris.target

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo
#max_depth=3 limita la profundidad para evitar sobreajuste y mantener el árbol interpretable.
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train, y_train)

# Predecir y calcular precisión
y_pred = dt.predict(X_test)
print(f"Precisión del Árbol de Decisión: {accuracy_score(y_test, y_pred):.2f}")

# Visualizar el árbol
plt.figure(figsize=(12, 8))
plot_tree(dt, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Árbol de Decisión para Iris")
plt.show() #permite ver cómo se entrena el árbol
```

##3. Bosques Aleatorios
### Teoría Básica
Un Bosque Aleatorio mejora los Árboles de Decisión cuando se utiliza muestreo con reemplazo (**bagging**) lo que permite entrenar varios árboles relacionados
Por otra parte es posible seleccionar subconjuntos aleatorios de características en cada división así como también se pueden combinar predicciones, Reduciendo el sobreajute y mejorando la generalización.



```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo
rf = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
rf.fit(X_train, y_train)

# Predecir y calcular precisión
y_pred_rf = rf.predict(X_test)
print(f"Precisión del Bosque Aleatorio: {accuracy_score(y_test, y_pred_rf):.2f}")
```
